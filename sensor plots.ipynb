```python
# Generate simulated temperature readings from two sensors.
# Sensor A: mean 25 C, std 3 C, 200 readings.
# Sensor B: mean 27 C, std 4.5 C, 200 readings.
# Also generate 200 timestamps uniformly from 0 to 10 seconds.
# Use np.random.default_rng with a seed = last 4 digits of your Drexel ID.


import numpy as np 
import matplotlib.pyplot as plt

seed = 3960 # replace with your Drexel ID's last 4 digits 
rng = np.random.default_rng(seed)

n = 200 
t = rng.uniform(0, 10, n) 
sensor_a = rng.normal(loc=25.0, scale=3.0, size=n)
sensor_b = rng.normal(loc=27.0, scale=4.5, size=n)

idx = np.argsort(t)
t, sensor_a, sensor_b = t[idx], sensor_a[idx], sensor_b[idx]

plt.figure(figsize=(8, 4))
plt.plot(t, sensor_a, '-o', markersize=4, label='Sensor A (µ=25, σ=3)')
plt.plot(t, sensor_b, '-s', markersize=4, label='Sensor B (µ=27, σ=4.5)')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (°C)')
plt.title('Simulated Temperature Readings')
plt.legend()
plt.grid(True)
plt.show()
```
